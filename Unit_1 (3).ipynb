## 24. Какая студия даёт самые длинные названия своим фильмам по количеству символов?

```python
import pandas as pd
movies = pd.read_csv('D:\Google\movie_bd_v5.csv')
movies['production_companies'] = movies['production_companies'].apply(lambda x: x.split(sep = '|'))
movies = movies.explode('production_companies')
movies['lenght'] = movies['original_title'].apply(lambda x: len(x))
final = movies.groupby(['production_companies'])['lenght'].max().sort_values()
display(final)
```

## 25. Описания фильмов какой студии в среднем самые длинные по количеству слов?

```python
import pandas as pd 
movies = pd.read_csv('D:\Google\movie_bd_v5.csv')
movies['total_words_in_overview'] = movies.overview.apply(lambda x: len(x.split(' ')))
movies['production_companies_list'] = movies['production_companies'].str.split('|')
movies_exploded = movies.explode('production_companies_list')
movies_exploded.groupby(['production_companies_list'])['total_words_in_overview'].mean().sort_values().idxmax()
```

## 27. Какие актеры чаще всего снимаются в одном фильме вместе?

```python
import pandas as pd
import collections
from itertools import combinations
movies = pd.read_csv('D:\Google\movie_bd_v5.csv')
movies['cast'] = movies['cast'].apply(lambda x:x.split('|'))
movies['pairs'] = movies['cast'].apply(lambda s: list(combinations(s, 2)))
movies = movies.explode('pairs')
collections.Counter(movies['pairs']).most_common(1)
```
